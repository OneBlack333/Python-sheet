```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
df = sns.load_dataset('titanic')
df.head(2)
```

```python
X = df[['pclass','age','sibsp']]
y = df['survived']
```

```python
X.fillna(df['age'].median(),inplace=True)
```

```python
X.isnull().sum()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.20,random_state=42)
```

```python
# Transformations
# Standardization : fit_transform = Train & transform = Test
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
X_train_scaled = scaler.fit_transform(X_train)
X_train_scaled
```

```python
X_test_scaled = scaler.transform(X_test)
X_test_scaled
```

```python
# we get this modeling datat afetr doing transformation
# Model Building
# fit() for training data & predict for test data
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
```

```python
lr.fit(X_train_scaled,y_train)
```

```python
lr.predict(X_test_scaled)
```
