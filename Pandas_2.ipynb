```python
import pandas as pd
import warnings
warnings.filterwarnings("ignore")
```

```python
df = pd.read_csv('Z:\Fortune_10.csv')
df.head(5)
```

```python
df.isnull().sum()
```

```python
df.fillna(1) # it fill all non values to 1
# limit 2 it show 1 value from 2 rows 
```

```python
# default it takes rows wise change used columns wise
df.fillna(method='ffill') # it take previous vlaue 
# there are some nan value because previous value not there 
```

```python
# {'backfill', 'bfill', 'ffill', None}
df.fillna(method='bfill') # it take forward vlaue to fill value
```

## Pandas Interpolate

# is used to fill NaN values in dataframes and series


```python
data = pd.read_csv('Z:\Excel_Practice\dataset excel\dataset.csv')
data
```

```python
data.interpolate(method='nearest') # different kinds of method parameter we can used
# this parameter used to fill nulll values 
# if we working on time series we can used 'quadratic'
# if we working on visualise the data or plot the data we can used 'akima'
```

```python
data.interpolate(limit=1)
# it fill column wise how many values u wanted to fill we can increase limit then that number of value get filled
```

```python
data.interpolate(limit=1,limit_direction='backward')
# default forward and by filling 1 value we can check how to filled
```

```python
data.interpolate(limit_area='inside')
```

```python
data.interpolate(limit_area='outside')
```
