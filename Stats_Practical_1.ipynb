## Stats Practical Implementation
### 

```python
import numpy as np
import statistics as st
import seaborn as sns
l1 = [22,33,44,55,66,21,32,43,45,61,21,30,40,50,60,100]
st.mode(l1)
```

```python
np.var(l1) #variance
```

```python
sns.histplot(l1,kde=True) # its not normal distribuitons
```

```python
# convert l1 dataset into normal distributions
# by using log method
sns.histplot(np.log(l1),kde=True)
```

```python
# for normal distribution random 1000 numbers
chart = np.random.normal(1,2,1000)
sns.histplot(chart,kde=True)
```

## lognormal dataset

```python
# for random dataset
# convert lognormal dataset into normal distribution
mean1,std1 = (2,1)
a = np.random.lognormal(mean1,std1,1000)
```

```python
sns.histplot(a,kde=True)
```

```python
sns.histplot(np.log(a),kde=True)
```

## This gives power law distribution

```python
a = 5
#1000 = shape
num = np.random.power(a,1000)

# This convert log normal dist to normal distribution
z = sns.histplot(np.log(num),kde=True)
```

## Q-Q PLOT

### If you want to check whether feature is guassian or normal distributed
### Q-Q plot

```python
import matplotlib.pyplot as plt
import scipy.stats as stat
import pylab
df = sns.load_dataset("iris")
```

```python
df.head(2)
```

```python
iris_data = sns.histplot(df['petal_length'],kde=True)
```

```python
def plot_data(sample):
    plt.figure(figsize=(10,6))
    plt.subplot(1,2,1)
    sns.histplot(sample,kde=True)
    plt.subplot(1,2,2)
    stat.probplot(sample,dist='norm',plot=pylab)
    plt.show()
```

```python
plot_data(df['sepal_length'])
```
